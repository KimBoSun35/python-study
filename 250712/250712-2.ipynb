```python
def bubble_sort(array):
    n = len(array)
    for i in range(n - 1):
        for j in range(n - i - 1):
            tmp = 0
            if array[j] > array[j + 1]:
                tmp = array[j]
                array[j] = array[j + 1]
                array[j + 1] = tmp
        print(array)
```

```python
#병원명 입력 시 전화번호 및 정보 제공(print) 구현
name=input("병원명: ")

result= df[df['병원명'].str.contains(name,case=False, na=False)]

if not result.empty:
    for idx, row in result.iterrows():
        print(f"\n병원명: {row['병원명']}")
        print(f"응급실 대표전화: {row['응급실 대표전화']}")
        print(f"도로명 주소: {row['도로명 주소']}")
        print(f"허가병상수: {row['허가병상수']}")
else:
    print("해당 병원을 찾을 수 없습니다.")
```

```python
#허가병상수 데이터 기반 평균 계산 및 상위 5개 병원 출력

#df안의 "허가병상수" 데이터는 현재 문자!(숫자로 변환 필요)
#평균계산(for문|numpy)
#상위 5개 병원 출력(fot문|numpy)
df['허가병상수'] = pd.to_numeric(df['허가병상수'], errors='coerce')
bed_list = df['허가병상수'].dropna().tolist()

total = 0
count = 0
for bed in bed_list:
    total += bed
    count += 1
avg_beds = total / count

print(f"전체 병원의 평균 허가병상수: {avg_beds:.2f}")
```
